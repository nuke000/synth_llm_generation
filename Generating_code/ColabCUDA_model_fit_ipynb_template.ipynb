```python
for i in range(1,traincycles+1) :
  pth = modelname+dataname+str(i*epoch_step)+'epochs.pth'
  print (pth)
  try:
    state = torch.load('/content/drive/MyDrive/DiplomaData/'+pth)
    last_pth = i*epoch_step
    last_i = i
    print ('- exists')
  except :
    print ('- na')
    break
print (modelname+dataname+str(last_pth)+'epochs.pth')
print (last_i)
```

```python
# Model fine-tunning cycle 

model = 1

for i in range(last_i, traincycles) :
  del model
  model = GReaT(llm=llm_path , batch_size=batch, epochs=epoch_step, save_steps=400000)
  #model.is_decoder=True #open for BERT
  #model.tokenizer.add_special_tokens({'pad_token': '[PAD]'}) #open for BERT
  if i > 0 :
    state = torch.load('/content/drive/MyDrive/DiplomaData/'+modelname+dataname+str(i*epoch_step)+'epochs.pth')
    model.model.load_state_dict(state)
    print (str(i*epoch_step)+' epochs model loaded')
  model.fit(traindata)
  torch.save(model.model.state_dict(), modelname+dataname+str((i+1)*epoch_step)+'epochs.pth')
  commandline = 'cp '+modelname+dataname+str((i+1)*epoch_step)+'epochs.pth /content/drive/MyDrive/DiplomaData'
  os.system(commandline)
  print (str((i+1)*epoch_step)+' epochs model saved')
```
